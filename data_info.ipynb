```python
import pandas as pd 
import json
import ast
```

```python
pd_char = pd.read_csv('characters.csv')
```

```python
#Count the lengths of the dict of each line in the column character and plot the distribution
pd_char['length'] = pd_char['characters'].apply(lambda x: len(eval(x)))
pd_char['length'].hist(bins=100)
```

```python
for i, row in pd_char.iterrows():
    M = 0
    F = 0
    print(row["characters"])
    res = ast.literal_eval(row["characters"])
    for gender in res.values():


        if gender == 'M':
            M += 1
        elif gender == 'F':
            F += 1
    pd_char.loc[i, 'M'] = int(M)
    pd_char.loc[i, 'F'] = int(F)
```
